# Q.1 Single Layer Perceptron

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron(object):
    def __init__(self, input_size, lr=1, epochs=100):
        # add one for bias
        self.W = np.zeros(input_size+1)
        self.epochs = epochs
        self.lr = lr
    
    def activation_fn(self, x):
        return 1 if x >= 0 else 0
 
    def predict(self, x):
        z = self.W.T.dot(x)
        a = self.activation_fn(z)
        return a
 
    def train(self, X, d):
        for _ in range(self.epochs):
            for i in range(d.shape[0]):
                # Bias inserted with input
                x = np.insert(X[i], 0, 1)
                y = self.predict(x)
                e = d[i] - y
                self.W = self.W + self.lr * e * x
            print("Iteration:", _,"Error: ",e)
            
    def test(self,t,w):
        for i in range(t.shape[0]):
            r = np.insert(t[i],0,w[0])
            p = w.T.dot(r)
            o = self.activation_fn(p)
            return o
```

```python
if __name__ == '__main__':
    X = np.array([
        [1, 1],
        [2, 2],
        [1, 6],
        [2, 7],
        [2, 4],
        [3, 1],
        [5, 5],
        [3, 2],
        [5, 6],
        [5, 7],
        [4, 2],
        [5, 1],
        [1, 3],
        [6, 4],
        [2, 4],
        [1, 4],
        [6, 5]
    ])
    d = np.array([0, 0, 1, 1, 1, 0, 1, 0, 1, 1, 0, 0, 0, 1, 0, 0, 1])
 
    perceptron = Perceptron(input_size=2)
    perceptron.train(X, d)
    print("The Final bias= ",perceptron.W[0]," and ", "The final weights= ",perceptron.W[1:])
    
    # Testing the points
    q = np.array([[6,3]])
    v = perceptron.test(q, perceptron.W)
    if v==0:
        print ("The tested points are below the Classification boundary.")
    else:
        print("The tested points are above the classification boundary.")
```
